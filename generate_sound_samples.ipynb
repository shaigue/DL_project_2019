## Creating a mono-channel Sine wave:
with the formula:

$sound[t]=sin(t\cdot a)$

```python
# tone synthesis
def sin_wave(freq, num_seconds=1, amplitude=1, sample_rate=44100):
    """ freq - the frecuancy of the sine wave,
        num_seconds - the length in seconds
        amplitude - the amplitude
        sample_rate - is the number of samples per second
    """
    # the length of the output array,  the 
    total_samples = num_seconds * sample_rate
    t = np.linspace(0,num_seconds, total_samples, dtype='float64')
    data = np.sin(2 * np.pi * freq * t) * amplitude
    return data
```
